```python
import json
import re
from datetime import datetime
import time # לצורך סימולציית זמן
import matplotlib.pyplot as plt
from google import genai
from google.genai.errors import APIError
from google.genai import types


GEMINI_API_KEY = " הכנס את המפתח שלך פה-"  

SERVICE_PROMPT = "יועץ טכני מקצועי וסבלני לבחירת מסלול לימודים. תפקידך הוא לשאול את המשתמש על כישוריו ונקודות החוזק שלו , להציע מסלולי לימוד מתאימים  ולתת ייעוץ מקצועי . טון הדיבור שלך צריך להיות אנושי ומתעניין, תומך, ומפרגן. אסור לך לחרוג מתחום מסלול הלימודים ואם יש סטיה עליך להגיד שאתה עונה רק בנושא של יעוץ תעסוקתי. השב בעברית בלבד."
# ציון < 0 (חיובי); ציון > 0 (שלילי)
SENTIMENT_DICT = {
    'מצוין': -2, 'נהדר': -1.5, 'מבוקש': -0.5, 'טוב': -1, 'מומלץ': -1.5, 'מרוצה': -2,
    'תודה': -1, 'אשמח': -0.5, 'חזק': -0.3,
    'יקר': 1.5, 'מעצבן': 2, 'תקלה': 3, 'בעיה': 2.5, 'אכזבה': 2, 'מבלבל': 1.5,
    'לא': 1, 'מסובך': 1.5, 'מאוכזב': 2.5, 'חסר': 1, 'גרוע': 3,
}
#אחר כך בגרפים ככל שהשיחה תהיה יותר שלילית כך הגרף יעלה יותר וככל שתהיה יותר חיובית אז בגרף ירד

# ----------------------------------------------------

class ServiceManager:
    """
    """
    def __init__(self, api_key: str, system_prompt: str, sentiment_dict: dict):
        """
        אתחול מנהל השירות.
        """
        self.api_key = api_key
        self.system_prompt = system_prompt
        self.sentiment_dict = sentiment_dict
        self.chat_history = [] 
        try:
            self.client = genai.Client(api_key=self.api_key)
            
            # יצירת אובייקט config המכיל את ה-system_instruction 
            config = types.GenerateContentConfig(
                system_instruction=self.system_prompt
            )

            
            self.chat = self.client.chats.create(
                model="gemini-2.5-flash", 
                config=config 
            )
            
            print("✅ הבוט מוכן לקבל את פני המשתמש. שירות 'יועץ תעסוקתי' הופעל.")
        except Exception as e:
            print(f"❌ שגיאת API/אתחול: אנא ודא שמפתח ה-API תקין ושהתקנת את הספרייה genai. שגיאה: {e}")
            self.client = None
            self.chat = None
    def _normalize_text(self, text: str) -> str:
        """מבצע נורמליזציה לטקסט: הסרת סימני פיסוק ויישור לאותיות קטנות."""
        # מנקה סימני פיסוק נפוצים
        text = re.sub(r'[.,!?;:()"]', '', text)
        return text.lower().strip()

    def calculate_sentiment(self, text: str) -> float:
        """
        מחשב את ציון הסנטימנט עבור מחרוזת טקסט נתונה.
        
        """
        normalized_text = self._normalize_text(text)
        words = normalized_text.split()
        
        # יצירת רשימת ציונים באמצעות List Comprehension
        scores = [self.sentiment_dict.get(word, 0) for word in words]
        
        return sum(scores)
        
    def _add_message_to_history(self, sender: str, text: str):
        """מוסיף הודעה עם ציון סנטימנט וחותמת זמן להיסטוריה."""
        score = self.calculate_sentiment(text)
        message = {
            'sender': sender,
            'text': text,
            'sentiment_score': score,
            'timestamp': time.time()
        }
        self.chat_history.append(message)

    def communicate_with_api(self, user_input: str) -> str:
        """
        מתקשר עם API Gemini, שולח את ההיסטוריה המלאה (דרך אובייקט ה-Chat), 
        ומטפל בשגיאות.
        """
        if not self.chat:
            return "מערכת הבוט אינה פעילה עקב שגיאת אתחול."

        # 1. שמירת הודעת המשתמש והסנטימנט שלה
        self._add_message_to_history(sender='User', text=user_input)
        
        try:
            # הבקשה ל-API כוללת את היסטוריית השיחה המלאה באופן אוטומטי
            response = self.chat.send_message(user_input)
            
            bot_response = response.text
            
            # 2. שמירת תגובת הבוט והסנטימנט שלה
            self._add_message_to_history(sender='Bot', text=bot_response)
            
            return bot_response
            
        except APIError as e:
            # טיפול בשגיאות תקשורת עם ה-API
            self.chat_history.pop() # הסרת הודעת המשתמש אם התקשורת נכשלה
            print(f"❌ שגיאת API במהלך שיחה: {e}")
            return "אירעה שגיאת תקשורת זמנית. אנא נסה שוב."
        except Exception as e:
            self.chat_history.pop()
            print(f"❌ שגיאה בלתי צפויה: {e}")
            return "אירעה שגיאה בלתי צפויה. אנא נסה שוב."


    # --- דרישות קלט/פלט וטיפול בקבצים ---

    def save_to_json(self, filename="chat_log.json"):
        """שומר את היסטוריית השיחה המלאה לקובץ JSON, כולל טיפול בשגיאות קובץ."""
        try:
            with open(filename, 'w', encoding='utf-8') as f:
                json.dump(self.chat_history, f, ensure_ascii=False, indent=4)
            print(f"\n💾 השיחה נשמרה בהצלחה לקובץ: **{filename}**")
            return filename
        except IOError as e:
            # טיפול בשגיאות קובץ (שמירה)
            print(f"❌ שגיאת שמירה לקובץ: {e}")
            return None

    def load_from_json(self, filename: str):
        """טוען היסטוריית שיחה מקובץ JSON קיים, כולל טיפול בשגיאות קובץ."""
        try:
            with open(filename, 'r', encoding='utf-8') as f:
                self.chat_history = json.load(f)
            print(f"✅ היסטוריית שיחה נטענה בהצלחה מקובץ: {filename}")
        except FileNotFoundError:
            print(f"❌ שגיאת קובץ: הקובץ '{filename}' לא נמצא.")
        except json.JSONDecodeError:
            print(f"❌ שגיאת קובץ: הקובץ '{filename}' אינו קובץ JSON תקין.")
        except IOError as e:
            # טיפול בשגיאות קובץ (טעינה)
            print(f"❌ שגיאת טעינה מקובץ: {e}")
            
    # --- דרישות עיבוד וניתוח נתונים (Analytics) ---

    def _get_sentiment_iterator(self, sender_filter: str = None):
        """
        Generator - דרישת advance python.
        מחזיר איטרטור (Generator) על ציוני הסנטימנט של הודעות רלוונטיות.
        """
        for message in self.chat_history:
            if sender_filter is None or message['sender'] == sender_filter:
                yield message['sentiment_score']
                
    def analyze_and_plot_results(self, filename: str):
        """טוען את הנתונים, מבצע ניתוח ומציג את הגרפים הנדרשים."""
        self.load_from_json(filename)

        if not self.chat_history:
            print("אין נתונים לניתוח או שהטעינה נכשלה.")
            return
        font_name = 'Miriam' 
        
        # 2. מגדיר את הגופן הזה כברירת מחדל לכל הטקסטים בגרף
        plt.rcParams['font.sans-serif'] = [font_name]
        plt.rcParams['font.family'] = 'sans-serif'
        
        # 3. מונע מ-matplotlib לקודד מחדש סימנים (כמו מינוס) שעלולים להיהרס בעקבות השינוי
        plt.rcParams['axes.unicode_minus'] = False

        # 1. ניתוח הסנטימנט לאורך השיחה
        user_messages = [msg for msg in self.chat_history if msg['sender'] == 'User']
        bot_messages = [msg for msg in self.chat_history if msg['sender'] == 'Bot']
        
        # 2. חישוב ממוצעים וציונים מצטברים
        user_scores = list(self._get_sentiment_iterator(sender_filter='User'))
        bot_scores = list(self._get_sentiment_iterator(sender_filter='Bot'))
        
        avg_user_sentiment = sum(user_scores) / len(user_scores) if user_scores else 0
        avg_bot_sentiment = sum(bot_scores) / len(bot_scores) if bot_scores else 0
        
        # ציון מצטבר (User Sentiment)
        cumulative_sentiment = [sum(user_scores[:i+1]) for i in range(len(user_scores))]

        # 3. ניתוח תדירות מילים שליליות
        
        # יצירת קבוצת המילים השליליות (ציון > 0) באמצעות Dictionary Comprehension
        negative_words_dict = {word: score for word, score in self.sentiment_dict.items() if score > 0}
        
        negative_word_count = {}
        for message in user_messages:
            text = self._normalize_text(message['text'])
            for word in text.split():
                if word in negative_words_dict:
                    negative_word_count[word] = negative_word_count.get(word, 0) + 1

        # 4. יצירת גרפים
        plt.style.use('ggplot')
        
        # --- גרף 1: מגמת שיחה (Line Chart) ---
        plt.figure(figsize=(10, 6))
        plt.plot(range(1, len(user_scores) + 1), cumulative_sentiment, marker='o', linestyle='-', color='darkblue')
        plt.axhline(0, color='gray', linestyle='--') # קו הנייטרליות
        plt.title(' מגמת שיחה: סנטימנט מצטבר של המשתמש לאורך השיחה')
        plt.xlabel('שלב בשיחה (הודעת משתמש)')
        plt.ylabel('ציון סנטימנט מצטבר (חיובי < 0, שלילי > 0)')
        plt.grid(True)
        plt.show()

        # --- גרף 2: השוואת טון (Bar Chart) ---
        plt.figure(figsize=(8, 5))
        senders = ['משתמש (User)', 'בוט (Bot)']
        averages = [avg_user_sentiment, avg_bot_sentiment]
        colors = ['red' if avg > 0 else 'green' for avg in averages]
        
        plt.bar(senders, averages, color=colors)
        plt.axhline(0, color='black', linestyle='-')
        plt.title(' השוואת טון: סנטימנט ממוצע - משתמש מול בוט')
        plt.ylabel('ציון סנטימנט ממוצע (חיובי < 0, שלילי > 0)')
        plt.show()

        # --- גרף 3: תדירות מילים שליליות (Bar Chart) ---
        if negative_word_count:
            plt.figure(figsize=(10, 6))
            words = list(negative_word_count.keys())
            counts = list(negative_word_count.values())
            
            # מיון מהשכיח ביותר
            sorted_data = sorted(zip(words, counts), key=lambda x: x[1], reverse=True)
            sorted_words, sorted_counts = zip(*sorted_data)
            
            plt.bar(sorted_words, sorted_counts, color='salmon')
            plt.title(' תדירות המילים השליליות ביותר של המשתמש')
            plt.xlabel('מילה שלילית')
            plt.ylabel('תדירות')
            plt.xticks(rotation=45, ha='right')
            plt.tight_layout()
            plt.show()
        else:
            print("אין מילים שליליות מובהקות בניתוח.")


# --- 3. לוגיקת הרצת הבוט ב-Notebook ---

if __name__ == "__main__":
    # יצירת מופע של המחלקה ServiceManager
    manager = ServiceManager(
        api_key=GEMINI_API_KEY, 
        system_prompt=SERVICE_PROMPT, 
        sentiment_dict=SENTIMENT_DICT
    )
    
    if manager.client is None:
        print("\nהפרויקט לא יכול להתחיל עקב כשל באתחול. אנא תקן את מפתח ה-API.")
    else:
        print("\n--- התחלת שיחה ---")
        print("אתה משוחח עם יועץ תעסוקתי ווירטואלי.")
        print("כדי לסיים את השיחה, הקלד: סיום")

        while True:
            user_input = input("👤 משתמש: ")
            
            if user_input.lower().strip() == "סיום":
                print("\n👋 סיום שיחה. מעבד נתונים...")
                break
                
            # תקשורת עם ה-API ושמירת ההיסטוריה באופן רציף
            bot_response = manager.communicate_with_api(user_input)
            print(f"🤖 בוט: {bot_response}")

        # 4. שמירה, טעינה וניתוח
        
        # (בונוס - שימוש ב-API לבקשת שם קובץ)
        try:
            chat_summary = "\n".join([f"{msg['sender']}: {msg['text']}" for msg in manager.chat_history])
            
            name_prompt = f"""בהתבסס על השיחה הבאה, הצע שם קובץ קצר ומתאים (באנגלית או בעברית) לתיעוד השיחה, 
בפורמט 'שם_קובץ_מתאים.json'. אנא החזר את שם הקובץ בלבד. השיחה:\n{chat_summary}"""
            
            name_response = manager.client.models.generate_content(
                model="gemini-2.5-flash",
                contents=name_prompt
            )
            
            # ניקוי שם הקובץ המוצע
            suggested_filename = name_response.text.strip().replace('"', '').replace("'", '').replace(' ', '_').replace('/', '_')
            
            if not suggested_filename.lower().endswith(".json"):
                 suggested_filename = f"{suggested_filename.replace('.json', '')}.json"
                 
            # שמירה עם השם שהוצע על ידי המודל
            final_filename = manager.save_to_json(suggested_filename)
            print(f"שם הקובץ המוצע על ידי ה-AI הוא: **{final_filename}**")
            
        except Exception:
            # אם ה-API נכשל בשלב זה, שומרים עם שם ברירת מחדל
            final_filename = manager.save_to_json("chat_log_default.json")
        
        # ניתוח והדמיית הנתונים
        if final_filename:
            print("\n--- ניתוח והצגת תוצאות ---")
            manager.analyze_and_plot_results(final_filename)
```
